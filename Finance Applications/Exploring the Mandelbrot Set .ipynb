```python
# (n) such that z -> z*z + n
# remains finite

# n=0 -> 0,0,0
# n=1 -> 0, 1, 2, 5, 26 -> explodes
# n = -1 -> 0, -1 -> does not explode (remains fintite)
# n = -2 -> remains finite
# n = -3 -> explodes

import numpy as np
```

```python
x, X = -2, 1
y, Y = -1, 1

delta = 0.001


re, im = np.mgrid[x:X:delta, y:Y:delta]
c = (re + 1j *im).reshape(im.shape[0], -1).T

z = np.zeros_like(c)
for i in range(100):
    z = z*z + c # mandelbrot eqn
    
figure(figsize=(20,10))
imshow(np.absolute(z), extent=(x,X,y,Y))
```

```python
x, X = -2, 1
y, Y = -1, 1

delta = 0.001


re, im = np.mgrid[x:X:delta, y:Y:delta]
c = (re + 1j *im).reshape(im.shape[0], -1).T

z = np.zeros_like(c)
escape = np.zeros_like(np.absolute(c))
for i in range(100):
    z = z*z + c # mandelbrot eqn
    idx = (np.absolute(z) > 4) & (escape ==0)
    escape[idx] = i
    
figure(figsize=(20,10))
imshow(escape, extent=(x,X,y,Y))
```

```python
def plot_mandelbrot(x, X, y, Y):
    delta = (X-x)/300
re, im = np.mgrid[x:X:delta, y:Y:delta]
c = (re + 1j *im).reshape(im.shape[0], -1).T

z = np.zeros_like(c)
escape = np.zeros_like(np.absolute(c))
for i in range(100):
    z = z*z + c # mandelbrot eqn
    idx = (np.absolute(z) > 4) & (escape ==0)
    escape[idx] = i
    
imshow(escape, extent=(x,X,y,Y))
```

```python
plot_mandelbrot(-2,1,-1,1)
```

```python
def plot_mandelbrot_at(x, y, D):
    return plot_mandelbrot(x-D, x+D, y-D, y+D)
```

```python
plot_mandelbrot_at(-.75,-.25,.1)
```
